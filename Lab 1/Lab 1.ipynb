# 1 . Data Frame

```python
#importing requitred libraries
import numpy as np
import pandas as pd
```

```python
#create pandas dataframe
myarray=np.array([[1,2,3],
                 [4,5,6]])

rownames=['a','b']

colnames=['f1','f2','f3']
```

```python
mydataframe=pd.DataFrame(myarray,index=rownames,columns=colnames)
print(mydataframe)
```

```python
#change the type of data
myarray=np.array([['a','sandhya',9.6],
                 [4,'shreya',6.5]])

rownames=['r1','r2']

colnames=['f1','f2','f3']
```

```python
mydataframe=pd.DataFrame(myarray,index=rownames,columns=colnames)
print(mydataframe)
```

# 6. Normalizing a column in pandas

```python
from matplotlib import pyplot as plt
```

```python
#Create a dataframe for a set of values in an array
myarray=np.array([1,2,-10,4,5,7,-4,-2,10])
mydataframe=pd.DataFrame(myarray)
print(mydataframe)
```

```python
#plot the data
mydataframe.plot(kind='bar')
plt.show()
```

```python
#Plot normalized data
from sklearn import preprocessing
fl_x=mydataframe.values.astype(float)
```

```python
#fl_x=mydataframe[['f1']].values.astype(float)

#if specific feature name is to be converted
```

```python
min_max_scaler=preprocessing.MinMaxScaler()
X_scaled=min_max_scaler.fit_transform(fl_x)
df_normalized=pd.DataFrame(X_scaled)
print(df_normalized)
```

```python
df_normalized.plot(kind='bar')
plt.show()
```
